```python
path="/content/drive/MyDrive/Colab Notebooks/ML/diabetes.csv"
import pandas as pd
import numpy as np
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
# support vector mission classifier
from sklearn import svm
from sklearn.metrics import accuracy_score
```

```python
data=pd.read_csv(path)
data.head()
```

```python
data.shape
```

Getting teh statistical measures of the data

```python
data.describe()
```

```python
data['Outcome'].value_counts()
```

```python
data.groupby('Outcome').mean()
```

Seperatring Outcome with data

```python
X=data.drop(columns='Outcome',axis=1)
# X=data.drop(0,axis=0) for rows
Y=data['Outcome']
```

Data Standardization

```python
scalar=StandardScaler()
```

```python
scalar.fit(X)
```

```python
standadized_data=scalar.transform(X)
```

```python
standadized_data
```

```python
X=standadized_data
X
```

Train Test Split

```python
# test_size: This parameter specifies the proportion of the dataset that should be allocated to the testing set. In this case, it's set to 0.2, which means 20% of the data will be used for testing, while the remaining 80% will be used for training.
# stratify: The stratify parameter is used for stratified sampling. When you set stratify=Y, it ensures that the class distribution in the original dataset is maintained in both the training and testing sets
# random_state: This parameter is used to seed the random number generator that splits the data into training and testing sets. Providing a value to random_state ensures that the split will be reproducible. If you use the same random_state value in the future, you'll get the same split. If you don't care about reproducibility, you can omit this parameter.
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,stratify=Y,random_state=2)
```

```python
X_test.shape
```

Training the model

```python
classifier=svm.SVC(kernel='linear')
```

Support vector machine classifier

```python
classifier.fit(X_train,Y_train)
```

Model Evaluation

Accuarcy score

Train Data

```python
X_train_accuarcy_prediction=classifier.predict(X_train)
training_data_accuracy=accuracy_score(X_train_accuarcy_prediction,Y_train)
```

```python
print(training_data_accuracy)
```

Test Data

```python
X_test_accuarcy_prediction=classifier.predict(X_test)
testing_data_accuracy=accuracy_score(X_test_accuarcy_prediction,Y_test)
print(testing_data_accuracy)
```

Making a predictive system

```python
input_data=[5,166,72,19,175,25.8,0.587,51]
# changing imported data as numpy array
input_data_as_numpy_array=np.asarray(input_data)
input_data_as_numpy_array=input_data_as_numpy_array.reshape(1,-1)
```

Standardize input data

```python
# For feature scaling, like in the case of the StandardScaler I mentioned earlier, you use .fit() to learn the scaling parameters (mean and standard deviation) from the training data. This ensures that the scaling is performed consistently with the distribution of the training data.

# When it comes to prediction, you don't generally use .fit() because you are not learning from the prediction data; instead, you are using a trained model to make predictions on new, unseen data

std_data=scalar.transform(input_data_as_numpy_array)

prediction=classifier.predict(std_data)
print(prediction)
```
